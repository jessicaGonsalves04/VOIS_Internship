```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("DoctorVisits.csv")
df.head()
```

```python
df.groupby("gender")["income"].sum().plot(kind="bar",title="Income Male Vs Female")
```

```python
df.groupby("gender")["illness"].sum().plot.bar(title="Total Illness in Male vs Female",ylabel="Sum of illness")
```

```python
#visits by male and female
df.groupby("gender")["visits"].count().plot(kind="pie",title="No of visits by male vs female",autopct="%1.0f%%")
```

```python
df.groupby("private")["gender"].count().plot.pie(autopct="%1.0f%%")
```

```python
female_df = df[df['gender'] == 'female']
female_private_counts = female_df.groupby("private").size().plot.pie(autopct="%1.0f%%")
```

```python
nchronic_df = df[df['nchronic'] == 'yes']
nchronic_counts = nchronic_df.groupby('gender').size().plot(kind='bar',title="Count of nChronic by Gender", xlabel="Gender", ylabel="Count")
```

```python
lchronic_df = df[df['lchronic'] == 'yes']
nchronic_counts = lchronic_df.groupby('gender').size().plot(kind='bar',title="Count of lChronic by Gender", xlabel="Gender", ylabel="Count")
```

```python
df.groupby("gender")["health"].sum().plot(kind="pie",title="Health male vs female",autopct="%1.0f%%")
```
